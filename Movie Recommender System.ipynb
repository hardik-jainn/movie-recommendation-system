```python
import pandas as pd
import numpy as np
import sys   #TO IMPORT sys.maxsize FOR PRINTING NUMPY ARRAYS WITHOUT TRUNCATION
import pickle #TO EXPORT THE LIST OF THE MOVIES TO PYCHARM FOR WEBSITE
```

```python
pd.__version__
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
'''NOW WE'LL CHECK FOR THE FEATURES THAT'D BE IMPORTANT FOR THE PURPOSE OF RECOMMENDATION
   THESE CAN BE :-
   Genres
   ID
   Keywords
   Overview
   Title
   Cast
   Crew'''
movies = movies[['genres','id','keywords','overview','title','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
'''AS YOU CAN SEE, OUTPUT IS A STRING THAT APPEARS TO BE A LIST OF DICTS.
   TO CONVERT THIS INTO A REAL LIST WE'LL USE ast.literal_eval '''
import ast
def convert_to_list(list_of_dictionaries):
    list_of_genres = []
    for i in ast.literal_eval(list_of_dictionaries):
        list_of_genres.append(i['name'])
    return list_of_genres
```

```python
movies['genres'] = movies['genres'].apply(convert_to_list)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert_to_list)
```

```python
movies.head()
```

```python
'''FOR GETING INFO OF THE TOP 3 CAST MEMBERS, WE'LL MAKE A NEW FUNCTION '''
def convert_cast(list_of_dictionaries):
    list_of_cast = []
    counter = 0
    for i in ast.literal_eval(list_of_dictionaries):
        if counter != 3:
            list_of_cast.append(i['name'])
            counter += 1
        else:
            break
    return list_of_cast
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
'''FOR GETING NAME OF THE DIRECTOR, WE'LL MAKE A NEW FUNCTION '''
def fetch_director(list_of_dictionaries):
    list_of_directors = []
    for i in ast.literal_eval(list_of_dictionaries):
        if i['job'] == 'Director':
            list_of_directors.append(i['name'])
            break
    return list_of_directors
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies_edited = movies[['id','title','tags']]
```

```python
movies_edited.head()
```

```python
movies_edited['tags'] = movies_edited['tags'].apply(lambda x :" ".join(x))
```

```python
movies_edited.head()
```

```python
movies_edited['tags'] = movies_edited['tags'].apply(lambda x : x.lower())
```

```python
movies_edited.head()
```

```python
'''NOW WE'LL PERFORM STEMMING TO AVOID DIFFERENT FORMS OF THE SAME WORD (FOR E.G. LOVE, LOVING, LOVED) '''
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_edited['tags'] = movies_edited['tags'].apply(stem)
```

```python
'''NOW WE'LL VECTORIZE MOVIES ON THE BASIS OF THE FREQUENCY OF
   EACH KEYWORD USED IN THEIR RESPECTIVE TAGS'''
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
'''5000 HERE REFERS TO THE NUMBER OF KEYWORDS WHOSE FREQUENCY
   WILL BE MEASURED TO VECTORIZE EACH MOVIE'''
```

```python
vectors = cv.fit_transform(movies_edited['tags']).toarray()
```

```python
vectors.shape # FOR 4806 MOVIES & 5000 KEYWORDS
```

```python
print(type(cv.get_feature_names_out()))
```

```python
with np.printoptions(threshold=sys.maxsize):
    print(cv.get_feature_names_out())
```

```python
'''EUCLIDEAN DISTANCE IN THIS CASE WON'T BE HELPFUL AS THE VECTORS 
   ORIGINATE FROM THE SAME POINT(ORIGIN)
   THEREFORE WE'LL CALCULATE COSINE(ANGULAR) DISTANCE '''
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
movies_edited[movies_edited['title']== 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    movie_index = movies_edited[movies_edited['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for indices in movies_list:
        print(movies_edited.iloc[indices[0]].title)
```

```python
recommend('Titanic')
```
